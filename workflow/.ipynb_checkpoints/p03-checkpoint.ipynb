Aim: To implement the Hill Climbing search algorithm and demonstrate its behaviour on a sample problem.

## Theory

Hill Climbing is a local search algorithm that continuously moves towards higher-value neighbours.
It does not maintain a search tree — only the current state is kept in memory.
The algorithm terminates when no neighbour has a higher value (a local maximum).
It can get stuck at local optima, ridges, or plateaus, which are its main limitations.

```python
import random

def objective(x):
    return -(x**2) + 10*x   # peak at x=5

def hill_climbing(start, step=0.5, max_iter=100):
    current = start
    for i in range(max_iter):
        neighbour = current + random.choice([-step, step])
        if objective(neighbour) > objective(current):
            current = neighbour
    return current

result = hill_climbing(start=0)
print(f"Best x found: {result:.2f}, f(x) = {objective(result):.2f}")
```

## Conclusion

The Hill Climbing algorithm was successfully implemented. The agent converges near the peak
of the objective function, demonstrating the greedy nature of local search. The experiment
also highlighted the risk of getting trapped at local optima depending on the starting point.
